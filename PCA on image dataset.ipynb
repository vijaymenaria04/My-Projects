```python
import numpy as np
from sklearn.decomposition import PCA
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open("nature.jpg")
image=np.array(image)
```

```python
data_reshaped= image.reshape(image.shape[0], -1)
```

```python
pca = PCA(n_components=100).fit(data_reshaped)
compressed = pca.transform(data_reshaped)
compressed_restored = pca.inverse_transform(compressed)
compressed_image = compressed_restored.reshape(image.shape)
```

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')
print(image.shape)
plt.subplot(1, 2, 2)
plt.imshow(compressed_image)
plt.axis('off')
plt.title('Compressed Image (100 components)')
print(compressed_image.shape)
plt.show()
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
total_explained_variance_ratio = np.sum(explained_variance_ratio)
print("Explained Variance Ratio:", total_explained_variance_ratio)
```
